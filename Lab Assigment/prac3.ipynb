```python
def factorial(n):
    if n == 0 or n == 1:   # base condition
        return 1
    else:
        return n * factorial(n - 1)

num = int(input("Enter a number: "))

if num < 0:
    print("Factorial is not defined for negative numbers")
else:
    print("Factorial of", num, "is:", factorial(num))
```

```python
print("Select an option:")
print("1. Probability of Coin Toss")
print("2. Probability of Dice Roll")

choice = int(input("Enter your choice (1 or 2): "))

if choice == 1:
    print("\nCoin Toss Selected")
    print("1. Head")
    print("2. Tail")
    
    coin_choice = int(input("Choose outcome (1 or 2): "))
    
    if coin_choice == 1 or coin_choice == 2:
        probability = 1 / 2
        print("Probability =", probability)
    else:
        print("Invalid coin choice")

elif choice == 2:
    print("\nDice Roll Selected")
    number = int(input("Enter the number (1 to 6): "))
    
    if 1 <= number <= 6:
        probability = 1 / 6
        print("Probability =", probability)
    else:
        print("Invalid dice number")

else:
    print("Invalid selection")
```

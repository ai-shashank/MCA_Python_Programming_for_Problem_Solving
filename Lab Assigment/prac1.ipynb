```python
# Pattern 1: Number Triangle
print("Pattern 1:\n")
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end="")
    print()

# small gap
print()

# Pattern 2: Centered triangular star pattern
print("Pattern 2:\n")
rows = 3
for i in range(1, rows + 1):
    # print a centered triangle using '* ' so stars are spaced nicely
    print(" " * (rows - i) + "* " * i)

print()

# Pattern 3: Bigger diamond pattern
print("Pattern 3:\n")
rows = 3
for i in range(1, rows + 1):
    print(" " * (rows - i) + "*" * (2 * i - 1))
for i in range(rows - 1, 0, -1):
    print(" " * (rows - i) + "*" * (2 * i - 1))

print()

# Pattern 4: Square border pattern
print("Pattern 4:\n")
n = 4
for i in range(n):
    for j in range(n):
        if i == 0 or i == n - 1 or j == 0 or j == n - 1:
            print("*", end="")
        else:
            print(" ", end="")
    print()

print()

# Pattern 5: Cross star pattern
print("Pattern 5:\n")
n = 3
for i in range(n):
    for j in range(n):
        if i == j or i + j == n - 1:
            print("*", end="")
        else:
            print(" ", end="")
    print()

print()
```
